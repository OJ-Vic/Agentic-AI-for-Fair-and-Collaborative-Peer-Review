````python
import os, json, glob, re, textwrap
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
````

````python
from datasets import load_dataset, Features, Value, Sequence
# =============================
# 1) Load PeerRead (local JSON) with explicit schema
# =============================
peerread_features = Features({
    "conference":         Value("string"),
    "VERSION":            Value("string"),
    "DATE_OF_SUBMISSION": Value("string"),
    "title":              Value("string"),
    "abstract":           Value("string"),
    "histories":          Sequence( Sequence(Value("string")) ),
    "reviews":            Sequence( Value("string") ),   # flat strings is OK for our workflow
    "SUBJECTS":           Value("string"),
    "authors":            Sequence(Value("string")),
    "accepted":           Value("bool"),
    "id":                 Value("string"),
    "COMMENTS":           Value("string"),               # appears in some files
})
 
DATA_FILES = {
    "train":      "PeerRead/data/arxiv.cs.ai_2007-2017/train/reviews/*.json",
    "validation": "PeerRead/data/arxiv.cs.ai_2007-2017/dev/reviews/*.json",
    "test":       "PeerRead/data/arxiv.cs.ai_2007-2017/test/reviews/*.json"
}
 
peerread = load_dataset("json", data_files=DATA_FILES, features=peerread_features)
df_train = peerread["train"].to_pandas()
 
# =============================
# 2) Build a robust paper_excerpt
# =============================
def build_excerpt(row):
    parts = []
    t = (row.get("title") or "").strip()
    a = (row.get("abstract") or "").strip()
    if t:
        parts.append(t)
    if a:
        parts.append(a)
    if not parts:
        # fallback to first non-empty review string
        rv = row.get("reviews")
        if isinstance(rv, (list, tuple)):
            rv = [r for r in rv if isinstance(r, str) and r.strip()]
            if rv:
                parts.append(rv[0].strip())
    return "\n\n".join(parts)[:2000]
 
df_train["paper_excerpt"] = df_train.apply(build_excerpt, axis=1)
df_train = df_train[df_train["paper_excerpt"].str.len().fillna(0) > 0].reset_index(drop=True)
````

````python
# =============================
# 0) LLM setup
# =============================
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
 
MODEL_ID = "Qwen/Qwen2.5-3B-Instruct"
 
# Use GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if device == "cuda" else torch.float32
 
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=True)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=torch_dtype,
    device_map="auto",              # moves it to GPU automatically if available
)
# ensure we have a pad token id to avoid warnings
if tokenizer.pad_token_id is None:
    tokenizer.pad_token_id = tokenizer.eos_token_id
    
def gen(prompt: str, max_new_tokens=280, temperature=1e-6, stop=None) -> str:
    """
    Wrap a plain prompt into Qwen's chat template and generate.
    Compatible with the rest of your agent code.
    """
    messages = [{"role": "user", "content": prompt}]
    input_ids = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to(model.device)
    greedy = (temperature is None) or (float(temperature) <= 0.0)

    gen_kwargs = dict(
        max_new_tokens=max_new_tokens,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.pad_token_id,
    )
    if greedy:
        gen_kwargs.update(do_sample=False)  # greedy decoding
    else:
        gen_kwargs.update(do_sample=True, temperature=float(temperature))
 
    with torch.no_grad():
        output_ids = model.generate(input_ids, **gen_kwargs)
 
    # decode only the newly generated continuation
    gen_ids = output_ids[0, input_ids.shape[-1]:]
    text = tokenizer.decode(gen_ids, skip_special_tokens=True)
 
    if stop:
        for s in stop:
            if s in text:
                text = text.split(s, 1)[0]
    return text.strip()
````

### Centralized Pipeline

````python
# =============================
# 3) Agent definitions
# =============================
class Reviewer:
    def __init__(self, name, style="balanced", trust=1.0):
        self.name = name
        self.style = style
        self.trust = trust
 
    def review(self, context: str) -> str:
        prompt = textwrap.dedent(f"""
        You are Reviewer {self.name}, a helpful, concise ACL-style reviewer.
        Write a structured, constructive review with sections:
        - Summary
        - Strengths (2 bullets)
        - Weaknesses (2 bullets)
        - Suggestions (1–3 concrete, actionable)
 
        Paper:
        \"\"\"{context}\"\"\"
 
        Review:
        """)
        return gen(prompt, max_new_tokens=320, temperature=0.4)
 
    def self_check(self, review_text: str) -> str:
        prompt = textwrap.dedent(f"""
        You are Reviewer {self.name}. Read your review and briefly refine it.
        Ensure clarity, fairness, and actionable suggestions. Return the improved review only.
 
        Review:
        \"\"\"{review_text}\"\"\"
        """)
        return gen(prompt, max_new_tokens=220, temperature=0.4)
 
    def critique(self, other_name: str, other_review: str, context: str) -> str:
        prompt = textwrap.dedent(f"""
        You are Reviewer {self.name}. Provide a short, professional critique of Reviewer {other_name}'s review.
        Focus on logical gaps, unsupported claims, and fairness. Be constructive.
 
        Paper:
        \"\"\"{context}\"\"\"
 
        Their review:
        \"\"\"{other_review}\"\"\"
 
        Your critique:
        """)
        return gen(prompt, max_new_tokens=180, temperature=0.5)
 
    def judge(self, review_text: str) -> dict:
        prompt = textwrap.dedent(f"""
        You are a meta-judge. Return ONLY a one-line JSON:
        {{"verdict": "Good" or "Not-Good", "constructiveness": 0..1}}
        Review:
        \"\"\"{review_text}\"\"\"
        JSON:
        """)
        text = gen(prompt, max_new_tokens=80, temperature=0.0, stop=["\n\n","Review:"])
        # --- robust parse ---
        m = re.search(r"\{.*\}", text, flags=re.S)
        if not m:
            return {"verdict": "Not-Good", "constructiveness": 0.5}
        try:
            j = json.loads(m.group(0))
            v = str(j.get("verdict","")).lower().replace(" ", "").replace("-", "")
            verdict = "Not-Good" if v == "notgood" else ("Good" if v == "good" else "Not-Good")
            c = float(j.get("constructiveness", 0.5))
            c = max(0.0, min(1.0, c))
            return {"verdict": verdict, "constructiveness": c}
        except Exception:
            return {"verdict": "Not-Good", "constructiveness": 0.5}
        m = re.search(r'(?i)"?constructiveness"?\D*([01](?:\.\d+)?|0?\.\d+)', text)
        c = float(m.group(1)) if m else 0.5
        return {"verdict": verdict, "constructiveness": max(0.0, min(1.0, c))}
````

````python
# 3 reviewers with slightly different styles (same power)
agents = [Reviewer("A","balanced",1.0), Reviewer("B","critical",1.0), Reviewer("C","supportive",1.0)]
 
# =============================
# 4) Centralized negotiation + trust-weighted vote
# =============================
def run_centralized(context: str, debate_rounds: int = 1) -> dict:
    # 1) initial self-consistency reviews
    reviews = {a.name: a.self_check(a.review(context)) for a in agents}
 
    # 2) argumentation-based negotiation (brief critique)
    critiques = {a.name: {} for a in agents}
    if debate_rounds > 0:
        names = [a.name for a in agents]
        for a in agents:
            for b in names:
                if b == a.name: 
                    continue
                critiques[a.name][b] = a.critique(b, reviews[b], context)
 
    # 3) trust scoring + local verdicts
    trust, verdicts = {}, {}
    for a in agents:
        j = a.judge(reviews[a.name])
        a.trust = 0.5 + 0.5 * j["constructiveness"]      # simple mapping → [0.5, 1.0]
        trust[a.name] = a.trust
        verdicts[a.name] = j.get("verdict", "Not-Good")
 
    # 4) trust-weighted vote
    w_good = sum(trust[n] for n,v in verdicts.items() if v == "Good")
    w_bad  = sum(trust[n] for n,v in verdicts.items() if v != "Good")
    decision = "Good" if w_good > w_bad else "Not-Good"
    reason = (
        f"Weighted votes — Good: {w_good:.2f}, Not-Good: {w_bad:.2f}. "
        f"Highest-trust agent: {max(trust, key=trust.get)}"
    )
 
    # 5) consolidated meta-review (self-consistency + human-ready)
    merge_prompt = textwrap.dedent("""
    You are the meta-reviewer. Merge the reviews below into one constructive meta-review.
    Mention consensus and key disagreements briefly, then give a final recommendation consistent with the vote: {decision}.
    ---
    """.format(decision=decision))
    all_reviews_block = "\n\n".join([f"[{k}] {v}" for k,v in reviews.items()])
    final_feedback = gen(merge_prompt + all_reviews_block, max_new_tokens=400, temperature=0.3)
 
    return {
        "decision": decision,
        "reason": reason,
        "final_feedback": final_feedback,
        "reviews": reviews,
        "critiques": critiques,
        "trust": trust,
        "verdicts": verdicts,
    }
````

````python
# =============================
# 5) Batch runner over a DataFrame
# =============================
def agentic_peer_review(df: pd.DataFrame, text_col="paper_excerpt", max_rows=3) -> pd.DataFrame:
    df = df.copy()
    rows = df[df[text_col].str.len().fillna(0) > 0].head(max_rows)
    out = []
    for _, r in rows.iterrows():
        res = run_centralized(r[text_col], debate_rounds=1)
        out.append({
            "id": r.get("id"),
            "title": r.get("title"),
            "agentic_decision": res["decision"],
            "agentic_reason": res["reason"],
            "agentic_meta_review": res["final_feedback"]
        })
    return pd.DataFrame(out)
 
# =============================
# 6) Run it
# =============================
MAX_ROWS = 50
df_results = agentic_peer_review(df_train, text_col="paper_excerpt", max_rows=MAX_ROWS)
df_results.head(10)
````

### Decentralised Pipeline

````python
class Reviewer:
    def __init__(self, name, style="balanced", trust=1.0):
        self.name = name
        self.style = style
        self.trust = trust
 
    def review(self, context: str) -> str:
        prompt = textwrap.dedent(f"""
        You are Reviewer {self.name}, a helpful, concise ACL-style reviewer.
        Write a structured, constructive review with sections:
        - Summary
        - Strengths (2 bullets)
        - Weaknesses (2 bullets)
        - Suggestions (1–3 concrete, actionable)
 
        Paper:
        \"\"\"{context}\"\"\"
 
        Review:
        """)
        return gen(prompt, max_new_tokens=320, temperature=0.4)
 
    def self_check(self, review_text: str) -> str:
        prompt = textwrap.dedent(f"""
        You are Reviewer {self.name}. Read your review and briefly refine it.
        Ensure clarity, fairness, and actionable suggestions. Return the improved review only.
 
        Review:
        \"\"\"{review_text}\"\"\"
        """)
        return gen(prompt, max_new_tokens=220, temperature=0.4)
 
    def critique(self, other_review: str, context: str = None) -> str:
        prompt = f"""You are {self.name}. Provide a short, constructive critique of the review below.
Review:
---
{other_review}
---
"""
        if context:
            prompt += f"\nPaper excerpt (for reference, do not summarize):\n---\n{context[:1200]}\n---\n"
        prompt += "Keep it under 120 words. Focus on reasoning gaps and missing evidence."
        return gen(prompt, max_new_tokens=160, temperature=0.4)
    def judge(self, review_text: str) -> dict:
        prompt = textwrap.dedent(f"""
        You are a meta-judge. Return ONLY a one-line JSON with keys:
        "verdict"  ("Good" or "Not-Good")
        "constructiveness" (number in [0,1])

        Review:
        \"\"\"{review_text}\"\"\"

        JSON:
        """).strip()

        txt = gen(prompt, max_new_tokens=80, temperature=0.0, stop=["\n\n","Review:"])

        # --- robust parse ---
        m = re.search(r"\{.*\}", txt, flags=re.S)
        if m:
            try:
                j = json.loads(m.group(0))
                v = str(j.get("verdict","")).lower().replace(" ", "").replace("-", "")
                verdict = "Not-Good" if v == "notgood" else ("Good" if v == "good" else "Not-Good")
                c = float(j.get("constructiveness", 0.5))
                c = max(0.0, min(1.0, c))
                return {"verdict": verdict, "constructiveness": c}
            except Exception:
                pass

        # fallback (no JSON): regex that cannot misread Not-Good as Good
        if re.search(r"(?i)\bnot[\s\-]*good\b", txt):
            verdict = "Not-Good"
        elif re.search(r"(?i)\bgood\b", txt):
            verdict = "Good"
        else:
            verdict = "Not-Good"

        m = re.search(r'(?i)constructiveness[^0-9]*([01](?:\.\d+)?|0?\.\d+)', txt)
        c = float(m.group(1)) if m else 0.5
        return {"verdict": verdict, "constructiveness": max(0.0, min(1.0, c))}
````

````python
import random, textwrap
from collections import defaultdict

agents = [Reviewer("A","balanced",1.0), Reviewer("B","critical",1.0), Reviewer("C","supportive",1.0)]

# ---------- helpers ----------
def _merge_reviews_to_meta(context, reviews_dict, decision):
    merge_prompt = textwrap.dedent(f"""
    You are the meta-reviewer for a peer-review committee.
    Paper excerpt:
    ---
    {context[:1400]}
    ---

    Below are committee member reviews (some might disagree). 
    Merge them into one constructive, concise meta-review (<= 220 words), 
    clearly noting major consensus and any key disagreements, and end with a
    final one-line recommendation consistent with the decision: {decision}.
    Reviews:
    ===
    {chr(10).join([f"[{k}] {v}" for k,v in reviews_dict.items()])}
    ===
    """)
    return gen(merge_prompt, max_new_tokens=320, temperature=0.3)

def _score_constructiveness(agent, review_text):
    try:
        j = agent.judge(review_text)
        # expected keys: 'verdict' in {"Good","Not-Good"}, 'constructiveness' in [0,1]
        return float(j.get("constructiveness", 0.5)), j.get("verdict", "Not-Good")
    except Exception:
        judge_prompt = textwrap.dedent(f"""
        Rate the following peer-review on two things:
        1) Binary paper verdict: "Good" or "Not-Good" (be decisive).
        2) Constructiveness score in [0,1] (0=unhelpful, 1=highly actionable).

        Review:
        ---
        {review_text}
        ---

        Respond strictly as JSON with keys verdict, constructiveness.
        """)
        jtxt = gen(judge_prompt, max_new_tokens=120, temperature=0.0)
        import json, re
        jtxt = re.sub(r"^```json|```$", "", jtxt).strip()
        try:
            j = json.loads(jtxt)
            return float(j.get("constructiveness", 0.5)), j.get("verdict", "Not-Good")
        except Exception:
            return 0.5, "Not-Good"
````

````python
# ---------- decentralized runner ----------
def run_decentralized_batch(
    df, 
    text_col="paper_excerpt", 
    id_col="id", 
    title_col="title",
    committee_size=3, 
    debate_rounds=1,
    trust_init=1.0, 
    trust_momentum=0.7,     # moving-average update
):
    """
    Decentralized: each paper -> small committee of agents.
    Returns a new DataFrame with agentic outputs.
    """
    # ensure at least committee_size agents
    assert len(agents) >= committee_size, "Need more agents for the chosen committee_size."

    # global trust per agent
    trust = {a.name: float(trust_init) for a in agents}

    rows = []
    for _, r in df.iterrows():
        context = str(r[text_col])
        paper_id = r.get(id_col, None)
        paper_title = r.get(title_col, "")

        # 1) choose committee
        committee = random.sample(agents, committee_size)

        # 2) self-consistency + initial review per agent
        reviews = {}
        for a in committee:
            base = a.review(context)
            reviews[a.name] = a.self_check(base)

        # 3) brief negotiation (pairwise critiques)
        if debate_rounds > 0:
            for _ in range(debate_rounds):
                for ai in committee:
                    for aj in committee:
                        if ai is aj:
                            continue
                        critique_text = ai.critique(reviews[aj.name], context)
                        reviews[aj.name] = reviews[aj.name] + f"\n\n[Critique from {ai.name}]: {critique_text}"

        # 4) compute local verdicts + update trust from constructiveness
        local_verdicts = {}
        for a in committee:
            cons, v = _score_constructiveness(a, reviews[a.name])
            local_verdicts[a.name] = v
            # update global trust (EMA)
            trust[a.name] = trust_momentum * trust[a.name] + (1 - trust_momentum) * cons

        # 5) trust-weighted vote
        w_good = sum(trust[a.name] for a in committee if local_verdicts[a.name] == "Good")
        w_bad  = sum(trust[a.name] for a in committee if local_verdicts[a.name] != "Good")
        total = w_good + w_bad
        decision = "Good" if w_good > 0.5 * total else "Not-Good"   # strict majority
        margin = abs(w_good - w_bad) / max(1e-6, total)

        reason = (
            f"Committee={','.join(a.name for a in committee)} | "
            f"Weighted votes — Good: {w_good:.2f}, Not-Good: {w_bad:.2f}. "
            f"Highest-trust voter: {max(committee, key=lambda a: trust[a.name]).name} "
            f"(trust={max(trust.values()):.2f}); margin={margin:.2f}."
        )

        # 6) meta-review
        meta = _merge_reviews_to_meta(context, reviews, decision)

        # 7) HITL flag (low-consensus)
        needs_human = (margin < 0.15)

        rows.append({
            "id": paper_id,
            "title": paper_title,
            "agentic_decision": decision,
            "agentic_reason": reason + (" [HITL recommended]" if needs_human else ""),
            "agentic_meta_review": meta,
            "committee": [a.name for a in committee],
            "local_verdicts": local_verdicts,
            "trust_snapshot": {a.name: round(trust[a.name], 3) for a in committee},
            "vote_margin": round(margin, 3),
        })

    return pd.DataFrame(rows)
````

````python
# Choose a small batch
MAX_ROWS = 50
df_small = df_train[df_train["paper_excerpt"].str.len() > 200].head(MAX_ROWS).copy()

df_decentral = run_decentralized_batch(
    df_small,
    text_col="paper_excerpt",
    id_col="id",
    title_col="title",
    committee_size=3,     # 3 reviewers per paper
    debate_rounds=1       # one short critique pass
)

df_decentral.head(10)
````

````python
import pandas as pd
import numpy as np
import re
import json
from textwrap import shorten

# df = pd.reaf_csv("peerread_agentic_reviews_centralised.csv")
df = pd.read_csv("peerread_agentic_reviews_decentralized.csv")
print(df.shape, df.columns.tolist()[:12])
````

````python
NUM = r'([0-9]+(?:\.[0-9]+)?)'

def parse_weights(reason_text: str):
    """Return (w_good, w_bad) as floats or NaN if missing."""
    if not isinstance(reason_text, str):
        return (np.nan, np.nan)

    g = re.search(rf'(?i)\bGood\b\s*[:=]\s*{NUM}', reason_text)
    b = re.search(rf'(?i)\bNot[\s\-]*Good\b\s*[:=]\s*{NUM}', reason_text)

    g_val = float(g.group(1)) if g else np.nan
    b_val = float(b.group(1)) if b else np.nan
    return (g_val, b_val)

def add_margin(df, reason_col="agentic_reason"):
    out = df.copy()
    vals = out[reason_col].apply(parse_weights).tolist()
    out[["w_good", "w_bad"]] = pd.DataFrame(vals, index=out.index)
    out["vote_margin"] = out["w_good"] - out["w_bad"]
    return out

# usage
df_hitl = add_margin(df)
MARGIN = 0.35
to_human = (df_hitl["vote_margin"].abs() < MARGIN) | df_hitl["vote_margin"].isna()
candidates = df_hitl.loc[to_human].copy()
print(f"Flagged for human review: {len(candidates)} of {len(df_hitl)}")
````

### Human review loop (accept/reject/skip + optional meta-review edits)

````python
from IPython.display import display, Markdown

def human_review_loop(
    df_candidates: pd.DataFrame,
    text_col="paper_excerpt",
    title_col="title",
    id_col="id",
    k=None  # limit how many to do this session
):
    df_candidates = df_candidates.copy()
    if k is not None:
        df_candidates = df_candidates.head(k)

    # Ensure storage columns exist
    for col in ["human_decision", "human_notes", "human_meta_review", "final_decision", "final_meta_review"]:
        if col not in df_candidates.columns:
            df_candidates[col] = np.nan

    reviewed_rows = []
    for idx, row in df_candidates.iterrows():
        title = row.get(title_col, "")
        pid   = row.get(id_col, "")
        excerpt = row.get(text_col, "")
        meta   = row.get("agentic_meta_review", "")
        reason = row.get("agentic_reason", "")
        decision = row.get("agentic_decision", "")

        print("\n" + "="*90)
        print(f"[{pid}] {title}")
        print(f"Agentic decision: {decision}")
        print(f"Reason: {reason}")
        print("\n--- PAPER EXCERPT (first 1200 chars) ---")
        print(shorten(str(excerpt).replace("\n"," "), width=1200, placeholder="…"))
        print("\n--- AGENTIC META-REVIEW ---")
        print(meta[:2000])
        print("\nActions: [A]ccept  [R]eject  [S]kip  [E]dit meta-review  [Q]uit")
        while True:
            ans = input("Your choice: ").strip().lower()
            if ans in ("a","r","s","e","q"):
                break

        if ans == "q":
            break
        if ans == "s":
            continue

        human_notes = input("Optional notes (ENTER to skip): ").strip()

        if ans == "e":
            print("\nPaste your revised meta-review. Finish with ENTER:")
            new_meta = input().strip()
            df_candidates.at[idx, "human_meta_review"] = new_meta if new_meta else np.nan
            # ask final human decision too
            ans2 = input("Final human decision [A]ccept/[R]eject: ").strip().lower()
            while ans2 not in ("a","r"):
                ans2 = input("Please type A or R: ").strip().lower()
            ans = ans2

        human_decision = "Good" if ans == "a" else "Not-Good"
        df_candidates.at[idx, "human_decision"] = human_decision
        df_candidates.at[idx, "human_notes"] = human_notes

        # Merge to final_* (human overrides)
        df_candidates.at[idx, "final_decision"] = human_decision
        df_candidates.at[idx, "final_meta_review"] = (
            df_candidates.at[idx, "human_meta_review"]
            if pd.notna(df_candidates.at[idx, "human_meta_review"])
            else row.get("agentic_meta_review", "")
        )

        reviewed_rows.append(idx)

    print(f"\nCompleted {len(reviewed_rows)} human reviews this session.")
    return df_candidates

# Run the loop on a small batch first. k = None for all flagged items
df_hitl_batch = human_review_loop(candidates, k=5)
````

### Update Trust with Human Decision

````python
def safe_json_loads(x):
    if isinstance(x, dict):
        return x
    if not isinstance(x, str) or not x.strip():
        return {}
    try:
        return json.loads(x)
    except Exception:
        return {}

def update_trust_with_human(df_in, trust_col="trust", verdicts_col="verdicts",
                            human_col="human_decision", eta=0.25):
    """
    For each paper with a human_decision:
      - If an agent's verdict == human_decision -> reward (+eta)
      - Else -> penalize (-eta/2)
    Trust values clamp to [0, 3] (arbitrary cap).
    """
    df_in = df_in.copy()
    if trust_col not in df_in.columns or verdicts_col not in df_in.columns:
        print("No trust/verdicts columns found; skipping trust update.")
        return df_in

    new_trust_col = f"{trust_col}_updated"
    df_in[new_trust_col] = None

    for idx, row in df_in.dropna(subset=[human_col]).iterrows():
        trust_map = safe_json_loads(row.get(trust_col))
        verdicts  = safe_json_loads(row.get(verdicts_col))
        if not trust_map or not verdicts:
            continue
        target = row[human_col]
        for agent, v in verdicts.items():
            if agent not in trust_map:
                continue
            if v == target:
                trust_map[agent] = min(3.0, trust_map[agent] + eta)        # reward
            else:
                trust_map[agent] = max(0.0, trust_map[agent] - eta/2.0)    # mild penalty
        df_in.at[idx, new_trust_col] = json.dumps(trust_map)

    return df_in

df_trust_updated = update_trust_with_human(df_merged)
````

### Evaluation

````python
import os, json, glob
import pandas as pd
from pathlib import Path
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix

DATA_ROOT = r"PeerRead/data/arxiv.cs.ai_2007-2017"

def load_peerread_labels(data_root=DATA_ROOT, splits=("train","dev","test")) -> pd.DataFrame:
    rows = []
    for split in splits:
        for fp in glob.glob(os.path.join(data_root, split, "reviews", "*.json")):
            with open(fp, "r", encoding="utf-8") as f:
                try:
                    j = json.load(f)
                except Exception:
                    continue
            paper_id = Path(fp).stem            # e.g., "0804.2155"
            lbl = j.get("accepted", None)
            if isinstance(lbl, bool):
                rows.append({
                    "id": paper_id,
                    "accepted": lbl,
                    "title": j.get("title", ""),
                    "abstract": j.get("abstract", ""),
                    "split": split
                })
    df = pd.DataFrame(rows).drop_duplicates(subset=["id"])
    return df
````

````python
def _norm_id_series(s: pd.Series) -> pd.Series:
    return (s.astype("string")
              .str.strip()
              .str.replace(r"\.0$", "", regex=True)
              .str.replace("\u200b", "", regex=False))
    
def evaluate_predictions(pred_csv, labels_df, id_col="id", decision_col="agentic_decision", positive_word="Good"):
    # read preds and normalize id
    preds = pd.read_csv(pred_csv)
    preds[id_col] = _norm_id_series(preds[id_col])
    # normalize labels id too
    lab = labels_df.copy()
    lab[id_col] = _norm_id_series(lab[id_col])
    # if your preds may contain duplicate ids, keep the last (or agg)
    preds = preds.dropna(subset=[id_col]).drop_duplicates(subset=[id_col], keep="last")
    # Keep only rows that have labels
    df = preds.merge(labels_df[["id","accepted","split"]], on=id_col, how="inner", validate="many_to_one")
    # Map strings to 0/1
    y_true = df["accepted"].astype(int)
    y_pred = (df[decision_col].str.strip().str.lower() == positive_word.lower()).astype(int)
    # Metrics
    acc = accuracy_score(y_true, y_pred)
    pr, rc, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    cm = confusion_matrix(y_true, y_pred, labels=[1,0])
    out = {
        "n": len(df),
        "accuracy": acc,
        "macro_precision": pr,
        "macro_recall": rc,
        "macro_f1": f1,
        "confusion_matrix_[truePos,trueNegRows]": cm.tolist(),
        "by_split": df.groupby("split").apply(lambda g: accuracy_score(g["accepted"].astype(int),
                                                                      (g[decision_col].str.strip().str.lower()==positive_word.lower()).astype(int))).to_dict()
    }
    return out, df
````

````python
# 1) Build labels once
labels = load_peerread_labels(DATA_ROOT)
print("Labels:", labels.shape, labels['accepted'].value_counts(dropna=False).to_dict())

# 2) Evaluate CENTRALIZED predictions
central_csv = "peerread_agentic_reviews_centralised.csv"
central_metrics, central_merged = evaluate_predictions(central_csv, labels)
print("Centralized:", central_metrics)

# 3) Evaluate DECENTRALISED predictions
decent_csv = "peerread_agentic_reviews_decentralized.csv"
decent_metrics, decent_merged = evaluate_predictions(decent_csv, labels)
print("Decentralized:", decent_metrics)
````
